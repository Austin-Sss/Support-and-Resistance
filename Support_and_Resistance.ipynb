```python
import pandas as pd #work with data frame
import numpy as np # data arrays
import yfinance # source of data
from mpl_finance import candlestick_ohlc
import matplotlib.dates as mpl_dates # candle stick chart 
import matplotlib.pyplot as plt # dates


#plot the data
plt.rcParams['figure.figsize'] = [12,7]
plt.rc('font', size=14)
```

```python
ticker_symbol = 'AMD'
ticker = yfinance.Ticker(ticker_symbol)

start_date = '2021-11-01'
end_date = '2022-01-10'
df=ticker.history(interval='1d', start=start_date, end=end_date) #intervals daily chart

df['Date'] = pd.to_datetime(df.index)
df['Date'] = df['Date'].apply(mpl_dates.date2num) #convert dates to format we want

df = df.loc[:,['Date', 'Open', 'High', 'Low', 'Close']] #only colunms we want

df
```

```python
#create two functions to calculate if a level is support or resistance 

#support
def is_Support_Level(df, i):
  support = df['Low'][i] < df['Low'][i - 1] and df['Low'][i] < df['Low'][i + 1] and df['Low'][i +1] < df['Low'][i + 2] and df['Low'][i - 1] < df['Low'][i - 2]
  return support

# resistance
def is_Resistance_Level(df, i):
  resistance = df['High'][i] > df['High'][i - 1] and df['High'][i] > df['High'][i + 1] and df['High'][i +1] > df['High'][i + 2] and df['High'][i - 1] > df['High'][i - 2]
  return resistance

#value = empty list
levels = []
level_types = [] 

for i in range(2, df.shape[0] - 2):

  if is_Support_Level(df, i):
    levels.append((i, df['Low'][i]))
    level_types.append('Support')

  elif is_Resistance_Level(df, i):
    levels.append((i, df['High'][i]))
    level_types.append('Resistance')

def plot_levels(): #"DateFormatter" is not defined
  fig, ax = plt.subplots()
  candlestick_ohlc(ax, df.values, width=0.6, colorup ='green', colordown = 'red', alpha=0.8)
  date_format = mpl_dates.DateFormatter('%d %b %y')
  ax.xaxis.set_major_formatter(date_format)
  fig.autofmt_xdate()
  fig.tight_layout()


  for level, level_type in zip(levels, level_types): #zip allows iterate over two list at same time
    plt.hlines(level[1],
               xmin = df['Date'][level[0]],
               xmax = max(df['Date']),
               colors = 'blue')
    plt.text(df['Date'][level[0]], level[1], (str(level_type) + ': ' + str(level[1]) + ' '), ha='right', va='center', fontweight='bold', fontsize='x-small')
    plt.title('Support and Resistance levels for ' + ticker_symbol, fontsize=24, fontweight='bold')
    fig.show()

plot_levels()

 
```
